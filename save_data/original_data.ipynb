```python
import sys
sys.path.append("..")
from src.load_data import load_dataframe_from_s3, save_dataframe_to_s3
from src.helpers import load_config_from_yaml
import pandas as pd
```

```python
# Define config file path
conf_file_path = "../config/local.yaml"
# Load config
yaml_conf = load_config_from_yaml(file_path=conf_file_path)

# Get bucket name and data name from config.
bucket_name = yaml_conf.get("BUCKET_NAME")

# Get all three batches
data_name = yaml_conf.get("BATCH1")
batch1_records = load_dataframe_from_s3(bucket_name, data_name)
data_name = yaml_conf.get("BATCH2")
batch2_records = load_dataframe_from_s3(bucket_name, data_name)
data_name = yaml_conf.get("BATCH3")
batch3_records = load_dataframe_from_s3(bucket_name, data_name)
```

```python
# Combine three batches into one
full_train_set = pd.concat([batch1_records, batch2_records, batch3_records],ignore_index=True)
len(full_train_set)
```

```python
cols_to_change = ['ER_STATUS_1','PR_STATUS_1','HER2_STATUS_1','ER_STATUS_2','PR_STATUS_2','HER2_STATUS_2','ER_STATUS_3','PR_STATUS_3','HER2_STATUS_3','ER_STATUS_4','PR_STATUS_4','HER2_STATUS_4']
for col in cols_to_change:
    print(full_train_set[col].value_counts())
```

```python
for col in cols_to_change:
    full_train_set[col] = full_train_set[col].map({"P":"positive","N":"negative","B":"borderline","X":"not performed","U":"unknown"})
full_train_set
```

```python
for col in cols_to_change:
    print(full_train_set[col].value_counts())
```
