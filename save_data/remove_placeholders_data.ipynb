```python
import sys
sys.path.append("..")
from src.load_data import load_dataframe_from_s3, save_dataframe_to_s3
from src.helpers import load_config_from_yaml
from src.data_prep import remove_placeholders
```

```python
# Define config file path
conf_file_path = "../config/local.yaml"
# Load config
yaml_conf = load_config_from_yaml(file_path=conf_file_path)

# Get bucket name and data name from config.
bucket_name = yaml_conf.get("BUCKET_NAME")
```

```python
str_cols = (
    [f"ER_SCORE_{i+1}" for i in range(4)] + 
    [f"PR_SCORE_{i+1}" for i in range(4)] + 
    [f"ER_STATUS_{i+1}" for i in range(4)] + 
    [f"PR_STATUS_{i+1}" for i in range(4)] + 
    [f"HER2_STATUS_{i+1}" for i in range(4)] + 
    [f"HER2ISH_STATUS_{i+1}" for i in range(4)]
)
```

```python
cleaned_data = load_dataframe_from_s3(bucket_name, "prompt_engineering_300_cleaned_reviewed.csv", str_cols)
```

```python
full_train_set_no_placeholders = cleaned_data.apply(remove_placeholders, col = "REPORT", axis = 1)
```
